```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
mjeseci = ['Sij', 'Velj', 'Ožu', 'Tra', 'Svi', 'Lip', 'Srp', 'Kol', 'Ruj', 'Lis', 'Stu', 'Pro']
dolasci = [1737, 5894, 13060, 30990, 43379, 94716, 123698, 120064, 60495, 16289, 3020, 2421]
nocenja = [9319, 17838, 35043, 97803, 150167, 447457, 786399, 864598, 357973, 55812, 12990, 9678]

# Create two separate figures for each subplot
fig1, ax1 = plt.subplots(figsize=(8, 4))
fig2, ax2 = plt.subplots(figsize=(8, 4))

# Plot data on the first subplot (line plot for dolasci)
ax1.plot(mjeseci, dolasci, marker='o', color='blue')
ax1.set_ylabel('Broj dolazaka')
ax1.set_title('Dolasci po mjesecima')
ax1.set_xticks(range(len(mjeseci)))  # Set x-axis ticks explicitly
ax1.set_xticklabels(mjeseci, rotation=45, ha='right')  # Set x-axis labels explicitly

# Annotate values with a slight offset to the right of the markers for the first subplot
for i, value in enumerate(dolasci):
    ax1.annotate(str(value), xy=(i, value), xytext=(6, 5), textcoords='offset points', ha='left', va='center',
                 arrowprops=dict(arrowstyle='-'))

# Plot data on the second subplot (line plot for nocenja)
ax2.plot(mjeseci, nocenja, marker='o', color='red')
ax2.set_ylabel('Broj noćenja')
ax2.set_title('Noćenja po mjesecima')
ax2.set_xticks(range(len(mjeseci)))  # Set x-axis ticks explicitly
ax2.set_xticklabels(mjeseci, rotation=45, ha='right')  # Set x-axis labels explicitly

# Annotate values with a slight offset to the right of the markers for the second subplot
for i, value in enumerate(nocenja):
    ax2.annotate(str(value), xy=(i, value), xytext=(6, 5), textcoords='offset points', ha='left', va='center',
                 arrowprops=dict(arrowstyle='-'))

# Manually adjust the layout to avoid overlapping
fig1.tight_layout()
fig2.tight_layout()

# Display the plots
plt.show()
```
